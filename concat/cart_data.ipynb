```python
import os
import pandas as pd
from typing import ByteString
```

```python
path = os.path.abspath("..")
```

```python
def save_csv_in_data_dir(df, file_name):
    df.to_csv(path+f"\data\{file_name}.csv", index=False)
    
def get_csv_file(name:str) -> pd.DataFrame:
    return pd.read_csv(f"{path}\data\{name}.csv")
```

```python
# generating demo cars sales csv file
data = {
    "car_id": [1, 2, 3, 4, 5],
    "model": ["Toyota Corolla", "Honda Civic", "Ford Focus", "Chevrolet Malibu", "Nissan Altima"],
    "year": ["2010.0", "2011.0", "2012.0", "2013.0", "NaM"],
    "sales": [200, 150, 300, 250, 400]
}
df = pd.DataFrame(data)
save_csv_in_data_dir(df, "cars_sales")
```

```python
data = {
    "car_id": [1, 2, 12, 13, 14],
    "model": ["Toyota Corolla", "Honda Civic", "Mercedes-Benz C-Class", "Audi A4", "Lexus IS"],
    "review_score": [4.2, 4.2, 3.9, "NaN", 4.0],
    "review_count": [150, 200, 60, 40, 70]
}
df = pd.DataFrame(data)
save_csv_in_data_dir(df, "cars_reviews")
```

```python
cars_sales = get_csv_file("cars_sales")
cars_reviews = get_csv_file("cars_reviews")
```

```python
# concat without any specification
pd.concat(objs=[cars_sales, cars_reviews])
```

```python
# with multiIndexing
combined_df=pd.concat(
    objs=[cars_reviews, cars_sales],
    keys= ["Sales", "Reviews"], #keys: sequence, default None
    ignore_index=True
)
combined_df
```

```python
# accessing price of Honda Civic
combined_df.loc[("Sales", 1)].iloc[3]
```

```python
# Horizontal Concatenation (Column-Wise Concatenation)
pd.concat(
    objs=[cars_sales, cars_reviews],
    axis=1
)
```
